## INI3 Strings and Lists
**Problem**: <br>
Given: A string s of length at most 200 letters and four integers a , b , c and d . <br>
Return: The slice of this string from indices a through b and c through d (with space in between), inclusively. In other words, we should include elements s[b] and s[d] in our slice.<br>
**Solution**:


```python
s = input()
a,b,c,d = map(int,input().split())
print(s[a:b+1],s[c:d+1])
```

## INI4 Conditions and Loops
**Problems**: <br>
Given: Two positive integers a and b (a<b<10000 ). <br>
Return: The sum of all odd integers from a through b , inclusively.<br>
**Solutions**:

```python
a,b = map(int, input().split())
total=0
for num in range(a,b+1):
    if num % 2 == 1:
        total += num

print (total)
```

## INI6 Intro to Python dictionary
**Problems**: <br>
Given: A string s of length at most 10000 letters. <br>
Return: The number of occurrences of each word in s, where words are separated by spaces. Words are case sensitive, and the lines in the output can be in any order. <br>
**Solutions**:

```python
s = input()
words = s.split()
record = {}
for word in words:
    if word in record:
        record[word]+=1
    else:
        record[word]=1

for key, value in record.items():
    print(key,"", value)
```
